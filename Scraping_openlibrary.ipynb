```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
web_url = "https://openlibrary.org/trending/yearly?page="

# Scraping 1 halaman
def scraping(no_halaman):
    url = f"{web_url}{no_halaman}"
    response = requests.get(url)

    # Status code
    if response.status_code != 200:
        print(f"Scraping gagal! {no_halaman}")
        return []

    # Parsing HTML
    soup = BeautifulSoup(response.text, 'html.parser')

    # Menemukan blok buku
    books = soup.find_all('div', class_='details')

    # Simpan data (list)
    data_buku = []

    # Loop ambil data tiap buku
    for book in books:
        # Judul
        title_tag = book.find('h3', class_='booktitle')
        title = title_tag.get_text(strip=True) if title_tag else 'N/A'

        # Penulis
        author_tag = book.find('span', class_='bookauthor')
        author = author_tag.get_text(strip=True).replace('by ', '') if author_tag else 'N/A'

        # Tahun pertama terbit
        published_tag = book.find('span', class_='publishedYear')
        first_published = published_tag.get_text(strip=True).replace('First published in ', '') if published_tag else 'N/A'

        # Jumlah edisi
        edition_tag = book.find('a', href=True, string=lambda x: 'edition' in x)
        edition = edition_tag.get_text(strip=True) if edition_tag else 'N/A'

        # Simpan data ke dictionary
        data_buku.append({
            'title': title,
            'author': author,
            'first_published': first_published,
            'edition': edition
        })

    return data_buku

# Crawling
def crawling(total_halaman):
    all_books = []
    for page in range(1, total_halaman + 1):
        print(f"Scraping halaman {page}... \n")
        page_books = scraping(page)
        all_books.extend(page_books)

    return all_books

# Crawling 10 halaman
books_data = crawling(10)

# Simpan ke CSV
df_books = pd.DataFrame(books_data)
df_books.to_csv('data_buku.csv', index=False)

# Output
for idx, book in enumerate(books_data, 1):
    print(f"{idx}. {book['title']} - {book['author']}, First Published: {book['first_published']}, Edition: {book['edition']}")
```
